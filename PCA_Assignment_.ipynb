#Data Processing

Normalising the data

```python
minmax=MinMaxScaler()
wine_norm=minmax.fit_transform(wine)
wine_norm
```

Standardise the data

```python
scaler=StandardScaler()
wine_std=scaler.fit_transform(wine)
wine_std
```

1)PCA on Normalised data

```python
pca=PCA()
pca_values=pca.fit_transform(wine_norm)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
#Variance captured by individual Principal Components
plt.figure(figsize=(12,7))
plt.bar(range(1,len(var)+1),var)

plt.title("Variance captured by individual Principal Components")
plt.xlabel("Number of components")
plt.ylabel("Variance(in %)")
plt.show()
```

```python
#Cumulative Variance
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
#Plot for the above cumulative variances
plt.figure(figsize=(12,7))
plt.plot(var1,color='green',linestyle='--',marker='o')

plt.title("Explained variance by each additional component")
plt.xlabel("Number of components")
plt.ylabel("Cumulative variance(in %)")
plt.show()
```

```python
pca=PCA(n_components=3)
pca_values=pca.fit_transform(wine_norm)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
#Cumulative Variance
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
#Creating a DataFrame with all cumulative PCA values for clustering
wine_norm_pca=pd.DataFrame(data=pca_values,columns=['PC0','PC1','PC2'])
wine_norm_pca
```

2)PCA on Standardised data

```python
pca=PCA()
pca_values=pca.fit_transform(wine_std)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
plt.figure(figsize=(12,7))
plt.bar(range(1,len(var)+1),var)

plt.title("Variance captured by individual Principal Components")
plt.xlabel("Number of component5s")
plt.ylabel("Variance(in %)")
plt.show()
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.figure(figsize=(12,7))
plt.plot(var1,color='blue',linestyle='solid',marker='o')

plt.title("Explained variance by each additional component")
plt.xlabel("No. of Components")
plt.ylabel("Cumulative Variance(in %)")
plt.show()
```

```python
pca=PCA(n_components=3)
pca_values=pca.fit_transform(wine_std)
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
wine_std_pca=pd.DataFrame(data=pca_values,columns=['PC0','PC1','PC2'])
wine_std_pca
```

#Clustering

##1) Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.metrics import silhouette_score as sil
```

**1.a) For Normalised PCA Data

```python
plt.figure(figsize=(20,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm_pca,method='complete'))
```

```python
plt.figure(figsize=(20,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm_pca,method='single'))
```

```python
plt.figure(figsize=(20,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm_pca,method='average'))
```

```python
plt.figure(figsize=(20,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm_pca,method='ward'))
```

ward linkage looks suitable because of the less distance between clusters but verify this with silhoutte score

```python
linkages=['complete','single','average','ward']

for i in range(2,6):
  for x in linkages:
    hc=AgglomerativeClustering(n_clusters=i,linkage=x)
    y_hc=hc.fit_predict(wine_norm_pca)
    sil_score=sil(wine_norm_pca,y_hc)
    print("For n:",i,"and linkage:",x,"Silhoutte score is:",np.round(sil_score,4))
```

```python
#As we predicted ward linkage to be best suitable it has vbeen proven the same so,
#Create clusters with n_clusters=3
hc=AgglomerativeClustering(n_clusters=3,linkage='ward',affinity='euclidean')
y_hc=hc.fit_predict(wine_norm_pca)
```

```python
wine1=pd.read_csv("wine.csv")
wine1['ClusterID_HC_norm']=y_hc
wine1[['Type','ClusterID_HC_norm']].head(15)
```

Cluster ID 2 --> Type 1

Cluster ID 0 --> Type 2

Cluster ID 1 --> Type 3

```python
wine1.groupby('ClusterID_HC_norm').agg(['mean'])
```

```python
wine1.groupby('Type').agg(['mean'])
```

Clusters made by algorithm and pre-defined clusters has negligible difference between them

```python
plt.figure(figsize=(20,8))
plt.scatter(wine_norm_pca.iloc[:,0],wine_norm_pca.iloc[:,1],c=wine1.ClusterID_HC_norm)

plt.title("PCA Plot")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

##2) K-Means Clustering

**2.a) For Normalised PCA Data**

```python
wcss=[]

for i in range(2,7):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(wine_norm_pca)
  wcss.append(kmeans.inertia_)

wcss
```

```python
plt.plot(range(2,7),wcss)
plt.title("Elbow Graph")
plt.xlabel("No. of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
for i in range(2,7):
  clusters=KMeans(n_clusters=i,random_state=0)
  clusters.fit(wine_norm_pca)
  sil_score=sil(wine_norm_pca,clusters.labels_)
  print("For Cluster ",i,", Silhoutte Score is ",sil_score)
```

Silhoutte score for 3 clusters is 0.47796 which is maximum among all, so we choose no. of clusters as 3.

```python
clusters=KMeans(n_clusters=3,random_state=1)
clusters.fit(wine_norm_pca)
```

```python
clusters.labels_
```

```python
wine1['ClusterID_kmeans_norm']=clusters.labels_
wine1[['Type','ClusterID_kmeans_norm']].head(15)
```

Cluster ID 1 --> Type 1

Cluster ID 2 --> Type 2

Cluster ID 0 --> Type 3

```python
wine1.groupby('ClusterID_kmeans_norm').agg(['mean'])
```

```python
wine1.groupby('Type').agg(['mean'])
```

Clusters made by algorithm and pre-defined clusters has negligible difference between them

```python
plt.figure(figsize=(20,8))
plt.scatter(wine_norm_pca.iloc[:,0],wine_norm_pca.iloc[:,1],c=wine1.ClusterID_kmeans_norm)

plt.title("PCA Plot")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
